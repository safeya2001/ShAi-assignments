# Data Cleaning

````python
missing_values = df.isnull().sum()
print("Missing values before imputation:")
print(missing_values)
````

````python
#I am  Filling missing values with the minimum value of the 'BasePay' column can be a conservative approach, assuming that missing values represent the lowest possible earnings.
missing_values_basepay = df['BasePay'].isnull().sum()
print("Missing values in 'BasePay' column before imputation:", missing_values_basepay)
min_basepay = df['BasePay'].min()
df['BasePay'] = df['BasePay'].fillna(min_basepay)
missing_values_basepay = df['BasePay'].isnull().sum()
print("Missing values in 'BasePay' column after imputation:", missing_values_basepay)
````

````python
missing_values_overtimepay = df['OvertimePay'].isnull().sum()
print("Missing values in 'OvertimePay' column before imputation:", missing_values_overtimepay)

# Fill missing values in the 'OvertimePay' column with zeros
df['OvertimePay'] = df['OvertimePay'].fillna(0)
missing_values_otherpay = df['OtherPay'].isnull().sum()
print("Missing values in 'OtherPay' column before imputation:", missing_values_otherpay)

# Fill missing values in the 'OtherPay' column with zeros
df['OtherPay'] = df['OtherPay'].fillna(0)
missing_values_overtimepay_after = df['OvertimePay'].isnull().sum()
print("Missing values in 'OvertimePay' column after zero fill:", missing_values_overtimepay_after)

missing_values_otherpay_after = df['OtherPay'].isnull().sum()
print("Missing values in 'OtherPay' column after zero fill:", missing_values_otherpay_after)
````

````python
#delete the 'Notes' and 'Status' columns because they are empty
missing_values_notes = df['Notes'].isnull().sum()
missing_values_status = df['Status'].isnull().sum()

print("Missing values in 'Notes' column:", missing_values_notes)
print("Missing values in 'Status' column:", missing_values_status)

# Drop the 'Notes' and 'Status' columns
df_cleaned = df.drop(columns=['Notes', 'Status'])

# Check the DataFrame after dropping the columns
print("\nDataFrame after dropping 'Notes' and 'Status' columns:")
print(df_cleaned.head())
````

````python
# fill missing values in the 'Benefits' column with zeros, assuming that missing values represent no benefits
missing_values_benefits = df['Benefits'].isnull().sum()
print("Missing values in 'Benefits' column before imputation:", missing_values_benefits)

# Fill missing values in the 'Benefits' column with zeros
df['Benefits'] = df['Benefits'].fillna(0)

# Check for missing values in the 'Benefits' column after imputation
missing_values_benefits_after = df['Benefits'].isnull().sum()
print("Missing values in 'Benefits' column after zero fill:", missing_values_benefits_after)
````

#Basic Data Visualization

````python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming 'df' is your DataFrame containing the salary and department data

# Visualization of salary distribution with a histogram
plt.figure(figsize=(10, 6))
plt.hist(df['TotalPay'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Total Pay')
plt.xlabel('Total Pay')
plt.ylabel('Frequency')

plt.grid(True)
plt.show()

# Visualization of proportion of employees in different departments with a pie chart
department_counts = df['JobTitle'].value_counts()
plt.figure(figsize=(20.,20))
plt.pie(department_counts, labels=department_counts , autopct='%1.1f%%', startangle=30)
plt.title('Proportion of Employees in Different Departments')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.show()
````

````python
# Visualization of BasePay distribution with a histogram
plt.figure(figsize=(10, 6))
plt.hist(df['BasePay'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Base Pay')
plt.xlabel('Base Pay')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

# Visualization of OvertimePay distribution with a histogram
plt.figure(figsize=(10, 6))
plt.hist(df['OvertimePay'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Overtime Pay')
plt.xlabel('Overtime Pay')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

# Visualization of OtherPay distribution with a histogram
plt.figure(figsize=(10, 6))
plt.hist(df['OtherPay'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Other Pay')
plt.xlabel('Other Pay')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
````

````python
plt.figure(figsize=(10, 6))
plt.hist(df['Benefits'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Benefits')
plt.xlabel('Benefits')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
````

#Grouped Analysis

````python
# Group the data by 'Year' and 'JobTitle' columns
grouped_data = df.groupby(['Year', 'JobTitle'])
summary_statistics = grouped_data['TotalPay'].agg(['count', 'mean', 'median', 'min', 'max'])

print("Summary statistics for each group:")
print(summary_statistics)

# Compare the average salaries across different groups
average_salaries = grouped_data['TotalPay'].mean()

print("\nAverage salaries across different groups:")
print(average_salaries)
````

#Simple Correlation Analysis


````python

import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

# Select the numerical columns to scale
numerical_columns = ['TotalPay', 'OtherPay']

# Scale the numerical columns using StandardScaler
scaler = StandardScaler()
df_scaled = pd.DataFrame(scaler.fit_transform(df[numerical_columns]), columns=numerical_columns)

# Check the correlation
correlation = df_scaled['TotalPay'].corr(df_scaled['OtherPay'])
print("Correlation between TotalPay and OtherPay (after scaling):", correlation)

plt.figure(figsize=(10, 6))
plt.scatter(df_scaled['TotalPay'], df_scaled['OtherPay'], color='skyblue', alpha=0.6)
plt.title('Scatter Plot: Total Pay vs Other Pay (after scaling)')
plt.xlabel('Total Pay (scaled)')
plt.ylabel('Other Pay (scaled)')
plt.grid(True)
plt.show()
````
